```python
df.tpep_dropoff_datetime = pd.to_datetime(df.tpep_dropoff_datetime)
df.tpep_pickup_datetime = pd.to_datetime(df.tpep_pickup_datetime)
```

```python
df['duration'] = df.tpep_dropoff_datetime - df.tpep_pickup_datetime
```

```python
td = df.duration.iloc[0]
```

```python
df.duration = df.duration.apply(lambda td: td.total_seconds() / 60)
```

```python
std_duration = df['duration'].std()
```

```python
# Q2 42.59
std_duration
```

```python
filtered_df = df[(df['duration'] >= 1) & (df['duration'] <= 60)]
```

```python
total_records = len(df)
records_after_filtering = len(filtered_df)
fraction_left = records_after_filtering / total_records
```

```python
# Q3 98
print("Fraction of records left after dropping outliers:", fraction_left)
```

```python
categorical = ['PULocationID', 'DOLocationID']
```

```python
df[categorical] = df[categorical].astype(str)
```

```python
df[categorical].describe()
```

```python
train_dicts = df[records_after_filtering].to_dict(orient='records')
```

```python
dv = DictVectorizer()
```

```python
dv.fit(train_dicts)
```

```python
# Q4
feature_matrix = dv.transform(train_dicts)

num_columns = feature_matrix.shape[1]
print("Dimensionality of the feature matrix (number of columns):", num_columns)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
X_train = dv.transform(train_dicts)
```

```python
X_train
```

```python
target_variable = 'duration'
y_train = df[target_variable].values
```

```python
y_train
```

```python
linear_reg = LinearRegression()
```

```python
linear_reg.fit(X_train, y_train)
```

```python
train_pred = linear_reg.predict(X_train)
```

```python
train_pred
```

```python
train_rmse = np.sqrt(mean_squared_error(y_train, train_pred))
```
